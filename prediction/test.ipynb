```python
import pandas as pd
import predict
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('../dataset/2022-01_AAPL.csv')

data = data[["Timestamp", "Volume"]]
data.head()

# create np ndarray for Volume data
input = data['Volume'].values
input = input[-336:]
input = np.expand_dims(input, axis=0)
input = np.expand_dims(input, axis=2)
print(input.shape)

args_path = "./results/PatchTST_AAPL_336_96/args.pkl"
model_path = "./results/PatchTST_AAPL_336_96/checkpoints/checkpoint.pth"

# predict
pred = predict.predict(model_path, args_path, input)

# plt

plt.plot(data['Volume'].values[-336:])
plt.plot(np.arange(336, 336+96), pred, color='red')
plt.show()
```
